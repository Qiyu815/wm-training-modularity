# Training progress analysis of Learning Brain data

The code below allows to visualize and analize data from 6-week working memory training study. Participants performed 18 sessions of dual n-back training (Jaeggi et al., 2008). Each session consisted of 20 runs of the task. The level of *n* increased with at least 80% correct trials in previous run.

```r
# testing training progress for experimental group

baseline <- lme(Nback_mean ~ 1, random = ~ 1 + Session|Subject, data = training_tidy_summary, method = 'ML',  control = list(opt = "optim"))
session <- update(baseline, .~. + Session)
session_quadr <- update(session, .~. + I(Session^2))
anova(baseline, session, session_quadr)
```

```r
#Place individual predictions and residuals into the dataframe
training_tidy_summary$pred_fl_rl_quadr <- predict(session_quadr)
training_tidy_summary$resid_fl_rl_quadr <- residuals(session_quadr)

#obtaining predicted scores for prototype
training_tidy_summary$proto_quadr <- predict(session_quadr, level=0)

p <- ggplot(training_tidy_summary, aes(x = Session, y = pred_fl_rl_quadr, group = Subject)) + 
    #geom_point(col = '#ee8c00', alpha = 0.3)+
    geom_line(col = '#ee8c00', alpha = 0.2, size = 1.5) +
    xlab('Training session') + 
    ylab('Predicted mean n level') +
    geom_line(aes(x = Session, y = proto_quadr), color='#262626ff',size=1.5, alpha = 0.5) +
    theme_training
 p  

ggsave("/home/finc/Dropbox/Projects/LearningBrain/Figures_NEW/Figure_NEW_pred_nback.pdf", plot = p, width = 10, height = 5, dpi = 300)  
```

Fitting Multilevel Model 
================================

a) Intercept only
-------------------

Setting baseline model which include only intercept.

```r
inter <- gls(nback_mean ~ 1,  
                 data = training_tidy_summary, 
                 method = 'ML', 
                 na.action = na.exclude)
summary(inter)
```

b) Unconditional Means Model 
--------------------------------------------------------
Fixed and random intercept only.

```r
fix_inter_rand_inter <- lme(fixed = nback_mean ~ 1, 
              random = ~ 1|sub, 
              data = training_tidy_summary,
              method = 'ML',
              na.action = na.exclude)

summary(fix_inter_rand_inter )
```

```r
# Extract the random effects with the ‘VarCorr()’ function
VarCorr(fix_inter_rand_inter)
```

```r
# We can compute the intra-class correlation (ICC) as the ratio of the random intercept 
# variance (between-person) to the total variance (between + within).

# First lets store the random effect variances, which will be the first column of the ‘VarCorr’ object.

random_effects <- as.numeric(VarCorr(fix_inter_rand_inter)[,1])
random_effects
```

```r
# Next lets compute the ICC. It is the ratio of the random intercept variance (between-person var) 
# over the total variance (between + within var)

ICC_between <- random_effects[1]/(random_effects[1]+random_effects[2]) 
ICC_between
```

- between-person variance = 59,4%;
- within-person variance = 40,6%

```r
# Place individual predictions and residuals from the unconditional means model um_fit into the dataframe
training_tidy_summary$pred_um <- predict(fix_inter_rand_inter)
training_tidy_summary$resid_um <- residuals(fix_inter_rand_inter)
head(training_tidy_summary)
```

```r
#plotting PREDICTED intraindividual change
#overlay PROTOTYPE (average individual)
#create the function for the prototype
fun_um <- function(x) {
  2.82186 + 0*x
}

p <- ggplot(training_tidy_summary, aes(x = session, y = pred_um, group = sub)) + 
    geom_line(col = '#ee8c00', alpha = 0.6, size = 1.5) +
    xlab('Training session') + 
    ylab('Predicted n-back level') +
    ggtitle("Unconditional Means Model") +
    theme_training +
    ylim(1,6) + 
    stat_function(fun = fun_um, color = "#262626ff", size = 2)
 p        

#ggsave("/home/finc/Dropbox/Projects/LearningBrain/fig_1b.pdf", plot = p, width = 5, height = 5, dpi = 300)  
```

```r
# Plotting residuals
p <- ggplot(training_tidy_summary, aes(x = session, y = resid_um, group = sub)) + 
    geom_line(col = '#ee8c00', alpha = 0.6, size = 1.5) +
    xlab('Training session') + 
    ylab('Residual n-back level') +
    ggtitle("Unconditional Means Model") +
    theme_training +
    ylim(-3,3) 

 p        
#ggsave("/home/finc/Dropbox/Projects/LearningBrain/fig_1b.pdf", plot = p, width = 5, height = 5, dpi = 300)  
```

c) Linear Growth Model
----------------------------------
Adding 'session' as a (time-varying) predictor. Examination of linear relation beween the time variable (session) and the outcome variable (nback_mean).

**Linear time model (MLM)**: random intercepts

My naming convention for objects here is: fixed linear (fl) and random intercept (ri)

```r
fix_lin_rand_inter <- lme(fixed = nback_mean ~ 1 + session, 
              random = ~ 1 |sub, 
              data = training_tidy_summary,              
              method = 'ML',
              na.action = na.exclude)

summary(fix_lin_rand_inter)
```

```r
#Place individual predictions and residuals into the dataframe
training_tidy_summary$pred_fl_ri <- predict(fix_lin_rand_inter)
training_tidy_summary$resid_fl_ri <- residuals(fix_lin_rand_inter)

#Create a function for the prototype
fun_fl_ri <- function(x) {
  2.1897840 + 0.0665343*x
}

p <- ggplot(training_tidy_summary, aes(x = session, y = pred_fl_ri, group = sub)) + 
    geom_line(col = '#ee8c00', alpha = 0.6, size = 1.5) +
    xlab('Training session') + 
    ylab('Predicted n-back level') +
    ggtitle("Fixed Linear, Random Intercept") +
    theme_training +
    ylim(1,6) + 
    stat_function(fun=fun_fl_ri, color="#262626ff", size = 2)
 p        
```

```r
# Plotting residuals
p <- ggplot(training_tidy_summary, aes(x = session, y = resid_fl_ri, group = sub)) + 
    geom_line(col = '#ee8c00', alpha = 0.6, size = 1.5) +
    xlab('Training session') + 
    ylab('Residual n-back level') +
    ggtitle("Fixed Linear, Random Intercept") +
    theme_training +
    ylim(-3,3)
 p   
```

f) Linear time model (MLM): modelling the covariance structure
-------------------------------------------------------

```r
fix_lin_rand_inter_slope_AR <- lme(fixed = nback_mean ~ 1 + session, 
              random = ~ 1 + session|sub, 
              data = training_tidy_summary,
              method = 'ML',
              correlation = corAR1(0, form = ~session|sub),
              na.action = na.exclude)
summary(fix_lin_rand_inter_slope_AR)
```

```r
intervals(fix_lin_rand_inter_slope_AR)
```

```r
#Place individual predictions and residuals into the dataframe
training_tidy_summary$pred_fl_rl_AR <- predict(fix_lin_rand_inter_slope_AR)
training_tidy_summary$resid_fl_rl_AR <- residuals(fix_lin_rand_inter_slope_AR)
#Create a function for the prototype
fun_fl_rl_AR <- function(x) {
  2.1762712 + 0.0673215*x
}

p <- ggplot(training_tidy_summary, aes(x = session, y = pred_fl_rl_AR, group = sub)) + 
    geom_line(col = '#ee8c00', alpha = 0.6, size = 1.5) +
    xlab('Training session') + 
    ylab('Predicted n-back level') +
    ggtitle("Fixed Linear, Random Linear, AR") +
    theme_training + 
    ylim(1,6) + 
    stat_function(fun=fun_fl_rl_AR, color="#262626ff", size = 2)
 p        
#ggsave("/home/finc/Dropbox/Projects/LearningBrain/fig_1b.pdf", plot = p, width = 5, height = 5, dpi = 300)  
```
